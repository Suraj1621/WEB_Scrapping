```python
from bs4 import BeautifulSoup
import requests
url='https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_nav_0 '
page=requests.get(url)
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
books=soup.find_all(class_='a-section a-spacing-none aok-relative')
len(books)
```

```python
first=books[0]
first
```

```python
bn=first.find(class_='p13n-sc-truncate p13n-sc-line-clamp-1').get_text().strip()
bn
```
